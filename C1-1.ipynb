```python
import pickle
import pandas as pd

with open("purchase_intent_model.pkl", "rb") as f:
    model = pickle.load(f)

test_users = pd.DataFrame([
    
    {
        'pages_viewed': 1,
        'cart_additions': 0,
        'has_cart': 0,
        'session_duration': 20,
        'engagement_score': 20,
        'past_purchase_count': 0,
        'is_returning_user': 0
    },

    
    {
        'pages_viewed': 3,
        'cart_additions': 0,
        'has_cart': 0,
        'session_duration': 120,
        'engagement_score': 360,
        'past_purchase_count': 0,
        'is_returning_user': 0
    },

    
    {
        'pages_viewed': 5,
        'cart_additions': 0,
        'has_cart': 0,
        'session_duration': 300,
        'engagement_score': 1500,
        'past_purchase_count': 1,
        'is_returning_user': 1
    },

    
    {
        'pages_viewed': 6,
        'cart_additions': 1,
        'has_cart': 1,
        'session_duration': 420,
        'engagement_score': 2520,
        'past_purchase_count': 2,
        'is_returning_user': 1
    },

    
    {
        'pages_viewed': 10,
        'cart_additions': 2,
        'has_cart': 1,
        'session_duration': 900,
        'engagement_score': 9000,
        'past_purchase_count': 5,
        'is_returning_user': 1
    }
])


probabilities = model.predict_proba(test_users)[:, 1]


def get_action(p):
    if p >= 0.7:
        return "High Intent – Target with offer"
    elif p >= 0.4:
        return "Medium Intent – Reminder campaign"
    else:
        return "Low Intent – No marketing spend"


results = test_users.copy()
results['purchase_probability'] = probabilities
results['recommended_action'] = results['purchase_probability'].apply(get_action)

results

```
